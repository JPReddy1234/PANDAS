```python
import pandas as pd

temperature_df=pd.DataFrame({"CITY":["mumbai","delhi","banglore","hyderabad"],"TEMPERATURE":[45,35,38,40]})
temperature_df
```

```python
humidity_df=pd.DataFrame({"CITY":["delhi","mumbai","banglore","chennai"],"HUMIDITY":[40,50,60,70]})
humidity_df
```

```python
pd.concat([temperature_df,humidity_df])
```

```python
pd.concat([temperature_df,humidity_df],ignore_index=True)
```

```python
pd.concat([temperature_df,humidity_df],axis=1)
```

```python
pd.concat([temperature_df,humidity_df],axis=0)
```

```python
pd.merge(temperature_df,humidity_df,how="outer")
```

```python
pd.merge(temperature_df,humidity_df,on="CITY",how="outer")
```
